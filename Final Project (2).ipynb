```python
#List of libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import linear_model
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#Load the data
file_path = 'Meat_prices.csv' #Dataset name
df = pd.read_csv(file_path, index_col='Month') #Hold the variable 'Month'
```

```python
#Clean the data
df = (df.replace({'%' : ''}, regex=True)) #Make all % values into trainable values
df = df.astype('float') #convert object variables into float
df = df.fillna(df.median()) #make all NaN values into median value

#Range of data
X = df.loc[:'20-Mar']
y = df.loc['May-90':] 
```

```python
#Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y,
test_size = 0.3,
random_state = 7)
```

```python
linreg = linear_model.LinearRegression() #Create linear regression model

linreg.fit(X_train,y_train) #fit the data using the training set

y_pred_train = linreg.predict(X_train) #predict on training set
y_pred_test = linreg.predict(X_test) #predict on test set

y_pred_train = pd.DataFrame(y_pred_train,columns=X.columns,index=y_train.index) #Convert prediction on training set to integer to plot
y_pred_test = pd.DataFrame(y_pred_test,columns=X.columns,index=y_test.index) #Convert prediction on test set to integer to plot

#Check mean squared error for every feature of training set
print("Mean Squared Error - Training Set")
print("Chicken Price", mean_squared_error(y_train['Chicken Price'],y_pred_train['Chicken Price']))
print("Chicken price % Change", mean_squared_error(y_train['Chicken price % Change'],y_pred_train['Chicken price % Change']))
print("Beef Price", mean_squared_error(y_train['Beef Price'],y_pred_train['Beef Price']))
print("Beef price % Change", mean_squared_error(y_train['Beef price % Change'],y_pred_train['Beef price % Change']))
print("Lamb Price", mean_squared_error(y_train['Lamb price'],y_pred_train['Lamb price']))
print("Lamb price % Change", mean_squared_error(y_train['Lamb price % Change'],y_pred_train['Lamb price % Change']))
print("Pork Price", mean_squared_error(y_train['Pork Price'],y_pred_train['Pork Price']))
print("Pork price % Change", mean_squared_error(y_train['Pork price % Change'],y_pred_train['Pork price % Change']))
print("Salmon Price", mean_squared_error(y_train['Salmon Price'],y_pred_train['Salmon Price']))
print("Salmon price % Change", mean_squared_error(y_train['Salmon price % Change'],y_pred_train['Salmon price % Change']))

#Check coefficient of determination (accuracy score) for every feature of training set
print("\nCoefficient Of Determination (Accuracy Score) - Training Set")
print("Chicken Price", abs(r2_score(y_train['Chicken Price'],y_pred_train['Chicken Price'])))
print("Chicken price % Change", abs(r2_score(y_train['Chicken price % Change'],y_pred_train['Chicken price % Change'])))
print("Beef Price", abs(r2_score(y_train['Beef Price'],y_pred_train['Beef Price'])))
print("Beef price % Change", abs(r2_score(y_train['Beef price % Change'],y_pred_train['Beef price % Change'])))
print("Lamb Price", abs(r2_score(y_train['Lamb price'],y_pred_train['Lamb price'])))
print("Lamb price % Change", abs(r2_score(y_train['Lamb price % Change'],y_pred_train['Lamb price % Change'])))
print("Pork Price", abs(r2_score(y_train['Pork Price'],y_pred_train['Pork Price'])))
print("Pork price % Change", abs(r2_score(y_train['Pork price % Change'],y_pred_train['Pork price % Change'])))
print("Salmon Price", abs(r2_score(y_train['Salmon Price'],y_pred_train['Salmon Price'])))
print("Salmon price % Change", abs(r2_score(y_train['Salmon price % Change'],y_pred_train['Salmon price % Change'])))

#Check mean squared error for every feature of test set
print("\nMean Squared Error - Test Set")
print("Chicken Price", mean_squared_error(y_test['Chicken Price'],y_pred_test['Chicken Price']))
print("Chicken price % Change", mean_squared_error(y_test['Chicken price % Change'],y_pred_test['Chicken price % Change']))
print("Beef Price", mean_squared_error(y_test['Beef Price'],y_pred_test['Beef Price']))
print("Beef price % Change", mean_squared_error(y_test['Beef price % Change'],y_pred_test['Beef price % Change']))
print("Lamb Price", mean_squared_error(y_test['Lamb price'],y_pred_test['Lamb price']))
print("Lamb price % Change", mean_squared_error(y_test['Lamb price % Change'],y_pred_test['Lamb price % Change']))
print("Pork Price", mean_squared_error(y_test['Pork Price'],y_pred_test['Pork Price']))
print("Pork price % Change", mean_squared_error(y_test['Pork price % Change'],y_pred_test['Pork price % Change']))
print("Salmon Price", mean_squared_error(y_test['Salmon Price'],y_pred_test['Salmon Price']))
print("Salmon price % Change", mean_squared_error(y_test['Salmon price % Change'],y_pred_test['Salmon price % Change']))

#Check coefficient of determination (accuracy score) for every feature of test set
print("\nCoefficient Of Determination (Accuracy Score) - Test Set")
print("Chicken Price", abs(r2_score(y_test['Chicken Price'],y_pred_test['Chicken Price'])))
print("Chicken price % Change", abs(r2_score(y_test['Chicken price % Change'],y_pred_test['Chicken price % Change'])))
print("Beef Price", abs(r2_score(y_test['Beef Price'],y_pred_test['Beef Price'])))
print("Beef price % Change", abs(r2_score(y_test['Beef price % Change'],y_pred_test['Beef price % Change'])))
print("Lamb Price", abs(r2_score(y_test['Lamb price'],y_pred_test['Lamb price'])))
print("Lamb price % Change", abs(r2_score(y_test['Lamb price % Change'],y_pred_test['Lamb price % Change'])))
print("Pork Price", abs(r2_score(y_test['Pork Price'],y_pred_test['Pork Price'])))
print("Pork price % Change", abs(r2_score(y_test['Pork price % Change'],y_pred_test['Pork price % Change'])))
print("Salmon Price", abs(r2_score(y_test['Salmon Price'],y_pred_test['Salmon Price'])))
print("Salmon price % Change", abs(r2_score(y_test['Salmon price % Change'],y_pred_test['Salmon price % Change'])))
```
