```python
#List of libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima_model import ARIMA
```

# Step 1: Load the data

```python
#Load the data
file_path = 'Meat_prices.csv' #Dataset name
df = pd.read_csv(file_path) #Hold the variable 'Month'
df.Month = pd.to_datetime(df.Month) #Make 'Month' into datetime
df.set_index('Month', inplace=True, drop=True) #Set 'Month' to index column
```

# Step 2: Clean the data

```python
#Clean the data
df = (df.replace({'%' : ''}, regex=True)) #Make all % values into trainable values
df = df.astype('float') #convert object variables into float
df = df.fillna(df.median()) #make all NaN values into median value
```

# Step 3: Check Moving Average, Rolling Mean/Standard Deviation and perform Dickey-Fuller Test for Meat Prices

```python
#Chicken Price
#Calculate Moving Average
df['cum_sum'] = df['Chicken Price'].cumsum()
df['count'] = range(1,len(df['Chicken Price'])+1)
df['mov_avg'] = df['cum_sum'] / df['count']

#Plot
plt.plot(df['Chicken Price'], 'b', label = "Chicken Price")
plt.plot(df['Chicken Price'].rolling(30).mean(), 'y', label = "Rolling Mean")
plt.plot(df['Chicken Price'].rolling(30).std(), 'k', label = "Rolling Standard Deviation")
plt.plot(df['mov_avg'], color = '0.6', label = "Moving Average")
plt.title("Chicken Price")
plt.xlabel("Year")
plt.ylabel("Price (dollars)")
plt.legend()
plt.show()

#Perform Dickey-Fuller test
result = adfuller(df['Chicken Price'])
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
	print('\t%s: %.3f' % (key, value))
```

# Step 7: Predict Meat Prices using ARIMA model

```python
#Predict Chicken Price using ARIMA model
AR_mod = ARIMA(df['Chicken Price'], order=(2,1,2)) 
AR_results = AR_mod.fit()

#Set up ARIMA prediction
AR_pred = pd.Series(AR_results.fittedvalues, copy=True)
AR_pred_cumsum = AR_pred.cumsum()
AR_log = pd.Series(df.iloc[-1], index=df.index)
AR_log = AR_log.add(AR_pred_cumsum,fill_value=0)
AR_prediction = np.exp(AR_log)

#Plot
plt.plot(df['Chicken Price'], 'b', label = "Chicken Price")
plt.plot(AR_prediction, 'k', label = "ARIMA Prediction")
plt.title("Chicken Price")
plt.xlabel("Year")
plt.ylabel("Price (dollars)")
plt.legend()
```
